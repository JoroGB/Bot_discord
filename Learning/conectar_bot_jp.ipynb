```python
class GetAcciones:
    def __init__(self, nombre, valor):
        self.nombre = nombre
        self.valor = valor

    @classmethod
    async def gekko(cls, argument):
        from pycoingecko import CoinGeckoAPI
        cg = CoinGeckoAPI()
        price = await cg.get_price(ids=argument, vs_currencies='usd')
        return cls(nombre=argument, valor=price[argument]['usd'])
```
